#  **File Processing for train data** 
# Here train data files are processed in the following cell. 4 individual band images are combined for each train image and stored in the ../output/train directory by year_month_n folders. Images are combined according to their labels.

```python
#Extract Labels for train data

labels = os.listdir('/kaggle/input/spacenus-ml-engineer-assignment/Spacenus_ML_Engineer_Assignment/Spacenus_ML_Engineer_Assignment/Field_boundary_detection_test/labels/train')
labels = [l[0:-4] for l in labels] 


#Extract and Create Output folders to save combined band images in train set

folders = os.listdir('/kaggle/input/spacenus-ml-engineer-assignment/Spacenus_ML_Engineer_Assignment/Spacenus_ML_Engineer_Assignment/Field_boundary_detection_test/train')

for folder in folders:
    output_dir_train = os.path.join("/kaggle/working/", "train")  # Update with your desired output directory
    output_dir = os.path.join(output_dir_train,folder)
    os.makedirs(output_dir, exist_ok = True)
    

    
# Define the path to the directory containing the band images
data_dir = "/kaggle/input/spacenus-ml-engineer-assignment/Spacenus_ML_Engineer_Assignment/Spacenus_ML_Engineer_Assignment/Field_boundary_detection_test/train"

# Define folders and labels (update with actual values)
#folders = ["year_month_1", "year_month_2", ...]  # Update with actual folder names
#labels = ["label_1", "label_2", ...]  # Update with actual label names

# List of spectral band names (B01, B02, B03, B04)
band_names = ["B01", "B02", "B03", "B04"]

# Loop through the folders and labels
for folder in folders:
    for label in labels:
        # Initialize an empty array to store the combined image
        combined_image = np.zeros((256, 256, 4), dtype=np.uint8)  # Assuming 256x256 is the image size

        # Loop through the band names and load and combine each band
        for i, band_name in enumerate(band_names):
            # Construct the file path for the current band image
            band_image_path = os.path.join(data_dir, folder, f"{label}_{band_name}.tif")

            # Load the band image
            band_image = cv2.imread(band_image_path, cv2.IMREAD_GRAYSCALE)

            # Normalize the pixel values to the range [0, 255]
            band_image = cv2.normalize(band_image, None, 0, 255, cv2.NORM_MINMAX)

            # Add the band to the combined image
            combined_image[:, :, i] = band_image

        # Define the path to save the combined image
        # Create the output directory if it doesn't exist
        output_dir = '/kaggle/working/train/' + f"{folder}/"
        output_filename = f"{label}.tif"
        output_path = os.path.join(output_dir, output_filename)

        # Print the output path for debugging
        #print(f"Saving combined image to: {output_path}")

        # Save the combined image
        cv2.imwrite(output_path, combined_image)

        # Display a message indicating that the image has been saved
        #print(f"Combined image saved to {output_path}")
```
